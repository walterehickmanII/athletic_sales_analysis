### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
data1 = pd.read_csv('Resources/athletic_sales_2020.csv')
data2 = pd.read_csv('Resources/athletic_sales_2021.csv')

combined_data = pd.concat([data1,data2], ignore_index=True)
# Handling missing values
combined_data.dropna(inplace=True) #Drop rows w/ missing values
#  Remove dups
combined_data.drop_duplicates(inplace=True)
combined_data.head()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
df1_reset = data1.reset_index()
df_merged = df1_reset.merge(data2, how='left')
print(df_merged)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
df_merged_group = pd.DataFrame(df_merged)
grouped_df = df_merged.groupby(['region', 'state', 'city'])['units_sold'].sum()
# Rename the sum to "Total_Products_Sold".
grouped_df = grouped_df.rename("Total_Products_Sold")
# Show the top 5 results.
top_5_results = grouped_df.nlargest(5).reset_index()
print(top_5_results)







                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
df_merged_pivot = pd.DataFrame(df_merged)

# Group by region, state, and city and calculate total products sold
grouped_df = df_merged.groupby(['region', 'state', 'city'])['units_sold'].sum()

# Rename the "units_sold" column to "Total_Products_Sold"
grouped_df = grouped_df.rename("Total_Products_Sold")

# Create a pivot table to show the top 5 results
pivot_table = grouped_df.reset_index().pivot_table(index=['region', 'state', 'city'],
                                                   values='Total_Products_Sold',
                                                   aggfunc='sum')

# Show the top 5 results
top_5_results = pivot_table.nlargest(5, 'Total_Products_Sold')

print(top_5_results)





```
